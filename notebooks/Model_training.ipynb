```python
import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv('data/gemstone.csv')

# Drop the 'id' column
df.drop(labels=['id'], axis=1, inplace=True)

# Separate dependent and independent features
X = df.drop(labels=['price'], axis=1)
y = df['price']

# Identify categorical and numerical columns correctly
categorical_col = X.select_dtypes(include='object').columns
numerical_col = X.select_dtypes(exclude='object').columns

# Define custom rankings for each ordinal variable
cut_categories = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

from sklearn.impute import SimpleImputer  # Handling missing values
from sklearn.preprocessing import StandardScaler  # Handling feature scaling
from sklearn.preprocessing import OrdinalEncoder  # Ordinal Encoding

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Numerical pipeline
num_pipe = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]
)

# Categorical pipeline
cat_pipeline = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ordinal_encoder', OrdinalEncoder(categories=[cut_categories, color_categories, clarity_categories])),
        ('scaler', StandardScaler())
    ]
)

# Combine numerical and categorical pipelines
processor = ColumnTransformer([
    ('numerical_pipeline', num_pipe, numerical_col),
    ('categorical_pipeline', cat_pipeline, categorical_col)
])

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=30)

# Fit and transform the data using the processor
X_train = pd.DataFrame(processor.fit_transform(X_train), columns=processor.get_feature_names_out())
X_test = pd.DataFrame(processor.transform(X_test), columns=processor.get_feature_names_out())

print(X_train.head())
print(X_test.head())
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Assuming X_train, X_test, y_train, y_test are already defined
# Fit a linear regression model as an example
regression = LinearRegression()
regression.fit(X_train, y_train)

def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, mse, rmse, r2_square

models = {
    'LinearRegression': LinearRegression(),
    'Lasso': Lasso(),
    'Ridge': Ridge(),
    'ElasticNet': ElasticNet()
}

trained_model_list = []
model_list = []
r2_list = []

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)  # Predict on the test set
    mae, mse, rmse, r2_square = evaluate_model(y_test, y_pred)  # Evaluate on the test set

    print(model_name)
    model_list.append(model_name)

    print('Model training performance:')
    print('RMSE:', rmse)
    print('MAE:', mae)
    print('R2 Score:', r2_square * 100)

    r2_list.append(r2_square)

    print('='*35)
    print('\n')
```
